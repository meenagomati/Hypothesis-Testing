```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats

import warnings
warnings.filterwarnings('ignore')
```

# Automobile Battery

```python
mu = 60
sigma = 6
xbar = 58
n = 40

#step 1:
#---------
#Mean lifetime of 40 randomly selected batteries is less than 58 months

#Step 2:
#---------
#Null hypothesis - current state of matter - Mean lifetime of randomly selected batteries is 58 months
#Alternate hypothesis - Mean lifetime of randomly selected batteries is less than 58 months

#Step 3:
#--------
#data ready , alpha = 0.05

#step 4:
#--------
#Finding the test - data follows normal distribution lets find the nrom test-static value

#step 5:
#-------
#Asuumptions are true sample size is 40(n>30) so it follows normal distribution

#Step 6:
#-------
#performing test

test_static = (xbar - mu) / (sigma / np.sqrt(n))
print('test_static : %.2f'%test_static)
p_value = norm.cdf(xbar, loc=mu, scale=sigma/np.sqrt(n))
print('p_value : %.2f'%p_value)
p_value

f,ax = plt.subplots(figsize=(10,7))
x1 = np.linspace(-4,4,50)
plt.plot(x1,norm.pdf(x1))
plt.axvline(x=test_static,color='red')
plt.show()
```

7.7e-08 is very much less than 0.05
p is very much less than alpha
null hypothesis is rejected
Mean lifetime of randomly selected batteries is less than 58 months

# **Electricity Confidence Interval**

```python
n = 40
mu = 310
sigma = 89

#finding out the confidence interval
confidence_interval = norm.interval(0.95, loc=mu, scale=sigma/np.sqrt(n))
print('confidence_interval : %.3f'%confidence_interval[0],confidence_interval[1])


test_stat1 = (mu - confidence_interval[0]) / (sigma / np.sqrt(n))
print('test_stat1 : %.2f'%test_stat1)
test_stat2 = (mu - confidence_interval[1]) / (sigma / np.sqrt(n))
print('test_stat2 : %.2f'%test_stat2)
```

```python
f,ax = plt.subplots(figsize=(10,7))
x1=np.linspace(-4,4,50)
plt.plot(x1,norm.pdf(x1))
plt.axvline(x=test_stat1,color='red')
plt.axvline(x=test_stat2,color='red')
plt.show()
```

# Jeans Company - Mean household income

```python
#step1 : question

#step 2: Framing hypothesis
#Null hypothesis - mean houshold income is less than or equal to 29000
#Alternate hypothesis - mean household income is greater than 29000

#step 3: setting alpha value as (0.05)
#step 5:
#Sample size greater than 30 & it follows normal distribution

mu = 29000
sigma = 8000
xbar = 30000
n = 400

#Step 6 - Performing test & drawing conclusion
z_stat = (xbar - mu) / (sigma / np.sqrt(n))
print('z_stat : %.2f'%z_stat)

#checking probability is greater than 29000
p_value = 1 - norm.cdf(z_stat)
print('p_value : %.4f'%p_value)

if(p_value < 0.05):
    print('Null hypothesis is accepted')
else:
    print('Null hypothesis is not rejected')

#print - Fact that mean household income is greater than 29000 is true
```
